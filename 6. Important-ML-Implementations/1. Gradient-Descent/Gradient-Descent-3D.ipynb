```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def z_function(x, y):
    return np.sin(5*x) * np.cos(5*y) / 5
```

```python
# calculating derivative
def calculate_gradient(x, y):
    dx = np.cos(5*x) * np.cos(5*y)
    dy = -np.sin(5*x) * np.sin(5*y)
    return dx, dy
```

```python
x = np.arange(-1, 1, 0.05)#-1 to 1 with stepsize of 0.05
y = np.arange(-1, 1, 0.05)
```

```python
X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)

current_pos = (0.7, 0.4, z_function(0.7, 0.4))
learning_rate = 0.01

ax = plt.subplot(projection="3d", computed_zorder = False)

for _ in range(1000):
    X_derivative, Y_derivative = calculate_gradient(current_pos[0], current_pos[1])
    X_new = current_pos[0] - learning_rate * X_derivative
    Y_new = current_pos[1] - learning_rate * Y_derivative
    Z_new = z_function(X_new, Y_new)
    current_pos = (X_new, Y_new, Z_new)

    ax.plot_surface(X, Y, Z, cmap="viridis", zorder = 0)
    ax.scatter(current_pos[0], current_pos[1], current_pos[2], color="red", zorder = 1)
    plt.pause(0.001)
    ax.clear()
```
