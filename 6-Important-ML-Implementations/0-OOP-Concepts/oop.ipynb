## 1. What is OOP?
Object Oriented Programming organizes code around objects (data + behavior). Python supports OOP through classes, objects, methods(functions inside class) and built-in methods.

### Why OOP?
- Organizes large codebases.
- Reusability via inheritance.
- Flexibility through polymorphism.
- Encapsulation for clearer, safer APIs.


## 2. Classes & Objects

### 2.1 Class
A class defines a blueprint. It defines:
- What data the objects will store (attributes).
- What actions the objects can perform (methods).


### 2.2 Objects
An object is an actual instance created from a class. For example if `class = blueprint` then object = the actual house built from the blueprint.
Objects have:
- Their own data (unique attribute values).
- Access to methods defined in the class.


```python
class Employee:
    def __init__(self):
        self.id = 1
        self.salary = 50000
        self.designation = "Software Engineer"
        
emp = Employee()

print(f"ID: {emp.id}")
```
