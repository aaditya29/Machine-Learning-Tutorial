# Building an RNN from Scratch in Python

### 1. Setup and Imports

```python
import time
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output

np.random.seed(42)
```

### 2. Prepare the Data

```python
text = """The only limit to our realization of tomorrow will be our doubts of today.
Let us move forward with strong and active faith."""
```

```python
text = text.lower()# Convert to lowercase for uniform processing
print(f"Corpus length: {len(text)} characters")
print(f"First 100 characters:\n{text[:100]}")
```

```python
chars = sorted(list(set(text)))  # Get unique characters in the text
char_to_idx = {ch: i for i, ch in enumerate(chars)}# Create a mapping from characters to indices
idx_to_char = {i: ch for i, ch in enumerate(chars)}# Create a mapping from indices to characters
vocab_size = len(chars)# Get the size of the vocabulary (number of unique characters)

print(f"\nTotal unique characters: {vocab_size}")
print(f"Characters: {''.join(chars)}")
```

### 3. Creating Training Sequences

```python
def create_sequences(text, seq_length=25):
    X = []
    y = []

    for i in range(0, len(text) - seq_length):
        # Input sequence
        sequence = text[i:i + seq_length]
        # Target character (the next character)
        target = text[i + seq_length]

        # Convert to indices
        X.append([char_to_idx[ch] for ch in sequence])
        y.append(char_to_idx[target])

    return np.array(X), np.array(y)


# Create sequences
seq_length = 25
X, y = create_sequences(text, seq_length)

print(f"Number of sequences: {len(X)}")
print(f"Input shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"\nExample sequence:")
print(f"Input:  '{''.join([idx_to_char[idx] for idx in X[0]])}'")
print(f"Target: '{idx_to_char[y[0]]}'")
```

```python
def one_hot_encode(sequence_indices, vocab_size):
    one_hot = np.zeros((len(sequence_indices), vocab_size))# Initialize a zero matrix for one-hot encoding
    one_hot[np.arange(len(sequence_indices)), sequence_indices] = 1# Set the appropriate indices to 1 for one-hot encoding
    return one_hot
```
