## Estimating Value of Pi Using Monte Carlo Simulation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pi_estimate(n=100000):#where n is the number of points randomly generated
    x = np.random.uniform(0,1, n)#generating n random points between 0 and 1
    y = np.random.uniform(0,1,n)#generating n random points between 0 and 1
    
    #Calculating distance from the circle origin of each point
    distances = np.sqrt(x**2 + y**2) 
    
    #Courting points inside the quarter circle
    points_inside = np.sum(distances <= 1)
    
    # Calculatint the value of pi
    pi_estimate = 4 * points_inside / n
    
    # Plotting the results of the point
    plt.figure(figsize=(10,10))
    plt.scatter(x[distances > 1], y[distances > 1], 
                color = 'red', alpha = 0.6, label = 'Outside')  
    plt.scatter(x[distances <= 1], y[distances <= 1],
                color = 'blue', alpha = 0.6, label = 'Inside')
    
    #Drawing the quarter circle
    theta = np.linspace(0, np.pi/2, 100)
    circle_x = np.cos(theta)
    circle_y = np.sin(theta)
    plt.plot(circle_x, circle_y, 'k-', linewidth=2)

    plt.axis('equal')
    plt.grid(True)
    plt.legend()
    plt.title(f'Monte Carlo Pi Estimation\nEstimated π = {pi_estimate:.6f}\nActual π = {np.pi:.6f}\nError = {abs(pi_estimate - np.pi):.6f}')
    plt.xlabel('x')
    plt.ylabel('y')

    return pi_estimate
```

```python
# Running simulation to calculate pi
estimated_pi = pi_estimate()
print(f"Estimated value of π: {estimated_pi:.6f}")
print(f"Actual value of π: {np.pi:.6f}")
print(f"Absolute error: {abs(estimated_pi - np.pi):.6f}")
```
