```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return 3*x**2 -4*x + 5
```

```python
f(3.0)
```

```python
xs = np.arange(-5, 5, 0.25)# -5 to 5 with step 0.25
ys = f(xs) # f(x) for each x in xs
plt.plot(xs, ys)
plt.title("Plot of the function f(x) = 3x^2 - 4x + 5")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

```python
h = 0.00001# small value for numerical derivative
x = 2/3# point at which we want to compute the derivative
(f(x + h) - f(x)) / h # numerical derivative at x
```

```python
a = 2.0
b = -3.0
c = 10.0
d = a*b + c
print(d)
```

```python
h = 0.0001

#taking inputs
a = 2.0
b = -3.0
c = 10.0

d1 = a*b + c
c += h# incrementing c by h
d2 = a*b + c# calculating new d with incremented c
print(f"Original d: {d1}, New d after incrementing c: {d2}")
print('slope:', (d2 - d1) / h) # slope is the change in d divided by change in c
```
